```python
# -*- coding: utf-8 -*-

# Compatibility between Python 2 and Python 3
from __future__ import division, print_function, unicode_literals, absolute_import

# Load libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
ar_0 = np.linspace(0, 0, num=100)
ar_a = np.linspace(0, 10, num=100)
ar_b = np.linspace(0, 5, num=100)
ar2 = np.array([ar_0, ar_a, ar_b])
```

```python
plt.plot(ar_a, '-.', label='Values')
plt.plot(ar2[1] - ar2[0], '--', label='Differences')
plt.plot(np.std(ar2, axis=0), label='Standard Deviation')
plt.plot(np.var(ar2, axis=0), label='Variance')
plt.legend()
```
