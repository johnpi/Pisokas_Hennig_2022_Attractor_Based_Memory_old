```python
# https://en.wikipedia.org/wiki/Leaky_integrator
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

t = np.linspace(0, 1, 1000)

tau = 0.1 # s at 1s the initial value would have dropped by 63% or to 36.8% of the initial value
y0 = 10
k = 1
y = k * np.exp(-t/tau) + y0
plt.plot(t, y)
```
